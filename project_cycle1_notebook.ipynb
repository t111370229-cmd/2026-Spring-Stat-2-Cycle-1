## 5. Sampling Simulation (Concept Preparation)

Goal: understand sampling variability of the sample mean.

TO DO:
1) Sample n = 30 observations (with replacement)  
2) Compute the sample mean of Glucose  
3) Repeat B = 1000 times  
4) Plot the distribution of sample means (matplotlib)  
5) Report the mean and standard deviation of the sampling distribution  


```python
# TO DO: set simulation parameters
B = 1000
n = 30
```

```python
means = []

for b in range(B):
    # TO DO: draw a bootstrap sample of size n (with replacement)
    sample = df.sample(n=n, replace=True)  # hint: df.sample
    
    # TO DO: compute mean of Glucose in this sample
    m = sample["Glucose"].mean()
    
    means.append(m)

means = np.array(means)
```

```python
plt.figure()

# TO DO: histogram of simulated sample means (hint: plt.hist)
plt.hist(means)

plt.title("Sampling Distribution of Mean(Glucose), n=30")
plt.xlabel("Sample mean of Glucose")
plt.ylabel("Count")
plt.show()
```

```python
# TO DO: report center and spread of the sampling distribution
# hint: means.mean() and means.std(ddof=1)
center = means.mean()
spread = means.std(ddof=1)
center, spread
```

Interpretation (5–8 lines):
- What does the sampling distribution represent?
- Is it narrower than the raw Glucose distribution? Why?
- How does this connect to uncertainty of an estimate?

### 6B. Extension Implementation (code + outputs)

```python
# TO DO: implement your extension here
import seaborn as sns
norm = pd.read_csv("norm.csv")
```

```python
norm_dict = {}

for col in norm.columns[1:]:  # 跳過 limit
    norm_dict[col] = norm[col].tolist()

print(norm_dict)
```

```python
def classify(value, thresholds):

    tiny, low, normal, high, excessive = thresholds

    if value <= tiny:
        return "Tiny"
    elif value <= low:
        return "Low"
    elif value <= normal:
        return "Normal"
    elif value <= high:
        return "High"
    elif value <= excessive:
        return "Excessive"
    else:
        return "Over"
classification_df = pd.DataFrame()

for col in norm_dict.keys():
    classification_df[col] = df[col].apply(
        lambda x: classify(x, norm_dict[col])
    )

print(classification_df.head())
```

```python
clean_df = df.copy()

mask = pd.Series(True, index=df.index)

for col in classification_df.columns:
    mask &= ~classification_df[col].isin(["Tiny","Over"])

clean_df = df[mask]

print("Before:", df.shape)
print("After:", clean_df.shape)
```

```python
import pandas as pd
import matplotlib.pyplot as plt

features = ["Glucose", "BloodPressure", "Insulin", "BMI"]

comparison_table = pd.DataFrame()

for col in features:
    comparison_table.loc[col, "Mean Before"] = df[col].mean()
    comparison_table.loc[col, "Mean After"] = clean_df[col].mean()

    comparison_table.loc[col, "Std Before"] = df[col].std()
    comparison_table.loc[col, "Std After"] = clean_df[col].std()

    comparison_table.loc[col, "Min Before"] = df[col].min()
    comparison_table.loc[col, "Min After"] = clean_df[col].min()

    comparison_table.loc[col, "Max Before"] = df[col].max()
    comparison_table.loc[col, "Max After"] = clean_df[col].max()

comparison_table = comparison_table.round(2)
print(comparison_table)

fig, axes = plt.subplots(2, 2, figsize=(16, 10))
axes = axes.flatten()

for i, feature in enumerate(features):
    values = comparison_table.loc[feature].values
    labels = comparison_table.columns.tolist()
    colors = ["skyblue" if "Before" in label else "salmon" for label in labels]

    bars = axes[i].bar(labels, values, color=colors)
    axes[i].set_title(f"{feature}: Before vs After Statistics")
    axes[i].set_ylabel("Value")
    axes[i].tick_params(axis='x', rotation=45)

    for bar in bars:
        height = bar.get_height()
        axes[i].text(
            bar.get_x() + bar.get_width() / 2,
            height,
            f"{height:.2f}",
            ha="center",
            va="bottom",
            fontsize=8
        )

plt.tight_layout()
plt.show()
```

```python
import seaborn as sns
sns.heatmap(df.corr(numeric_only=True), annot=True)
plt.title("Correlation heatmap before cleaning")
plt.show()
sns.heatmap(clean_df.corr(numeric_only=True), annot=True)
plt.title("Correlation heatmap after cleaning")
plt.show()
```

```python
features = ["Glucose","BloodPressure","Insulin","BMI"]

for col in features:

    value_counts = classification_df[col].value_counts()

    labels = value_counts.index
    sizes = value_counts.values

    plt.figure(figsize=(5,5))

    plt.pie(
        sizes,
        labels=labels,
        autopct='%1.1f%%',
        startangle=90
    )

    plt.title(f"{col} Value Distribution")

    plt.show()
```
